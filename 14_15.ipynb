```python
import pandas as pd
```

```python
filename = '/content/drive/MyDrive/FINAL_ACT1_PAM/movies_metadata.csv'
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df = df.rename(columns={'original_title': 'original_title',
'overview': 'overview',
'release_date	': 'release_date	'})
# DataFrame after renaming columns
df.head()
```

```python
df.columns =['', 'adult',	'belongs_to_collection',	'budget'	,'genres',	'homepage',	'id',	'imdb_id',	'original_language',	'original_title',	'overview',	'...',	'release_date',	'revenue',	'runtime'	,'spoken_languages',	'status',	'tagline',	'title',	'video',	'vote_average',	'vote_count','','']
df.head()
```

```python
df = df.drop(columns=['original_title'])
df.head()
```

```python
del df['belongs_to_collection']
df.head()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.dropna(subset=['original_language'], inplace=True)
df.isna().sum()
```

```python
df.drop(columns='overview', inplace=True)
df.isna().sum()
```

```python
df.dropna(axis=1, inplace=True)
```

```python
df.dropna(axis=1, thresh=.45, inplace=True)
```

```python
df['release_date'].fillna('Missing', inplace=True)
```

```python
median_age = df['status'].median()
df['status'].fillna(median_age, inplace=True)
df.isna().sum()
```

```python
most_common_genres = df['genres'].mode()
df['genres'].fillna(most_common_genres, inplace=True)
```
